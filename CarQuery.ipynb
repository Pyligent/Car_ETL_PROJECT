```python
import requests
import json
import pandas as pd
from pprint import pprint
import os
```

```python
file_path=os.path.join(os.environ["HOMEPATH"], 'Documents\Data Analysis Bootcamp\Car_ETL_PROJECT',"carquery.json")
with open(file_path) as f:
    json_string = f.read().replace('\n', '')[2:-2]
    data = json.loads(json_string)
```

```python
pprint(data)
```

```python
car_list=[]

for i in range(len (data["Trims"])):
    car_list.append([            
        data['Trims'][i]['model_year']
        ,data['Trims'][i]['make_display']
        ,data['Trims'][i]['model_transmission_type']
        ,data['Trims'][i]['model_lkm_hwy']
        ,data['Trims'][i]['model_engine_cc']
       ])
```

```python
car_df=pd.DataFrame(car_list)
```

```python
car_df.head()
```

```python
car_df_rename=car_df.rename(columns={0:"Year", 1:"model", 2:"Transmission", 3:"mpg", 4:"Displacement"})
```

```python
car_df_rename.head()
```

```python
car_df_rename.to_csv("carquery.csv", 
                  encoding="utf-8", index=False, header=True)
```
